```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv("/content/hacker-news-posts/HN_posts_year_to_Sep_26_2016.csv")
data.head()
```

```python
#1. В якому році та місяці було найбільше публікацій?
```

```python
data['created_at'] = pd.to_datetime(data['created_at'], format='%m/%d/%Y %H:%M')
grouped_data = data.groupby(data['created_at'].dt.date).size().reset_index(name='count')
fig = px.line(grouped_data, x='created_at', y='count', title='Number of Posts Over Time')
fig.update_xaxes(title_text='Date')
fig.update_yaxes(title_text='Number of Posts')
fig.show()
```

```python
data['year'] = data['created_at'].dt.year
data['month'] = data['created_at'].dt.month

post_counts = data.groupby(['year', 'month']).size().reset_index(name='count')
top = post_counts.sort_values(by='count', ascending=False).iloc[0]
top
```

```python
# найбільше публікацій було у 10 місяці (жовтні) 2015 року
```

```python
year = top['year']
month = top['month']

top_month_data = data[(data['year'] == year) & (data['month'] == month)]
```

```python
top_month_data
```

```python
#
```

```python
# Extract the date and weekday from the 'created_at' column
top_month_data['date'] = top_month_data['created_at'].dt.date
top_month_data['weekday'] = top_month_data['created_at'].dt.strftime('%A')

# Group by date and weekday, then calculate the count
grouped_data = top_month_data.groupby(['date', 'weekday']).size().reset_index(name='count')

# Plot the data
fig = px.line(grouped_data, x='date', y='count', color='weekday', title='Number of Posts Over Time by Weekday')
fig.update_xaxes(title_text='Date')
fig.update_yaxes(title_text='Number of Posts')
fig.show()
```

```python
# бачимо, що на вихідних (saturday, sunday) трафік найменший
```
