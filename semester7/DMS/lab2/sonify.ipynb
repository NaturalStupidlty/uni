```python
import lightkurve
import matplotlib.pyplot as plt

from astronify.series import SoniSeries
```

```python
kep12b_lc = (lightkurve.search_lightcurve('KIC 11804465', quarter=1).download_all()
             .stitch())
```

```python
f, ax = plt.subplots(figsize=(12, 6))
ax.plot(kep12b_lc['time'].jd, kep12b_lc['flux'])
ax.set_xlabel("Time (JD)")
ax.set_ylabel("Flux")

plt.show()
```

```python
kep12b_obj = SoniSeries(kep12b_lc)
kep12b_obj.sonify()
```

```python
kep12b_obj.play()
```

```python
kep12b_obj.stop()
```

```python
# export as MIDI file
kep12b_obj.write("kep12b.mid")
```
